# Workign with Large Data:-

```python
for df in pd.read_csv('./pandas-master/pokemon_data.csv', chunksize=5):
    print("CHUNK DF")
    print(df)
```

```python
new_df_chunk = pd.DataFrame(columns=df.columns)

for df in pd.read_csv('./pandas-master/pokemon_data.csv', chunksize=5):
    
    result = df.groupby(['Type 1']).count()

    new_df_chunk = pd.concat([new_df_chunk, result])
    
    print("CHUNK DF")
    print(df)
```
