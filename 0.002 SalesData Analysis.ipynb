# Using Apply and lambda :-

```python
# Finding city Using Lambda

All_Data['City'] = All_Data['Purchase Address'].apply(lambda i: i.split(',')[1])

All_Data.head()
```

```python
def get_city(address):
    return address.split(',')[1]

def get_state(address):
    return address.split(',')[2].split(' ')[1]

All_Data['City'] = All_Data['Purchase Address'].apply(lambda i: get_city(i))

All_Data['State'] = All_Data['Purchase Address'].apply(lambda i: get_state(i))

All_Data['City_State'] = All_Data['Purchase Address'].apply(lambda i: get_city(i) + '  (' + get_state(i) + ')')

# OR you can use string formatting 

#apply(lambda x: f"{get_city(x)}  ({get_state(x)})")


All_Data.head()
```

```python
City_Result = All_Data.groupby('City_State').sum()

City_Result
```

```python


#Cities = All_Data['City_State'].unique()   This may give you a wrong result because of order

Cities = [city for city, df in All_Data.groupby('City_State')]

plt.bar(Cities, City_Result["Sales"].map(int))


plt.xticks(Cities, rotation='vertical', size=8)     #For rotation

plt.title('City Wise Sales Report', fontsize = 18 )

plt.ylabel('Sales in USD ($)', size=12)
plt.xlabel('Cities', size=12)

plt.show()
```

# Which Products are often sold together???

```python
All_Data.head(4)
```

```python
df = All_Data[All_Data['Order ID'].duplicated(keep=False)]

df.head()
```

```python
df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda i: ','.join(i))

df.head()
```

```python
df = df[['Order ID' , 'Grouped']].drop_duplicates()

df.head(20)
```

# Count the pairs that occure together in Grouped Columns

```python
from itertools import combinations
from collections import Counter
```

```python
# Referenced: https://stackoverflow.com/questions/52195887/counting-unique-pairs-of-numbers-into-a-python-dictionary
from itertools import combinations
from collections import Counter

count = Counter()

for row in df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))
    #above 2 is showing 2 itoms together
 

for key ,value in count.most_common(10):
    print(key,value)
    
#count.most_common(10)
    
#print(count)
# Counter({('iPhone', 'Lightning Charging Cable'): 1005, 
#          ('Google Phone', 'USB-C Charging Cable'): 987, 
#          ('iPhone', 'Wired Headphones'): 447, 
#          ('Google Phone', 'Wired Headphones'): 414, 
#          ('Vareebadd Phone', 'USB-C Charging Cable'): 361, 
#          ('iPhone', 'Apple Airpods Headphones'): 360, 
#          ('Google Phone', 'Bose SoundSport Headphones'): 220,
```

# What Product Sold Most ? Why do you think it sold the most?

```python
All_Data.head()
```

```python
product_group = All_Data.groupby('Product')

product_group.sum()
```

```python
Quantity_Ordered = product_group.sum()['Quantity Ordered']

Quantity_Ordered
```

```python
products = [Product for Product , df in product_group]

products
```

```python
plt.bar(products,Quantity_Ordered)

plt.xticks(products, rotation='vertical', size=8)     #For rotation

plt.title('Products Wise Sale', fontsize = 18 )

plt.ylabel('Count', size=12)
plt.xlabel('Products', size=12)

plt.grid()

plt.show()
```

```python
prices = All_Data.groupby('Product').mean()['Price Each']



fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(products, Quantity_Ordered, color='g')
ax2.plot(products, prices, color='b')

ax1.set_xlabel('Product Name')
ax1.set_ylabel('Quantity Ordered', color='g')
ax2.set_ylabel('Price ($)', color='b')
ax1.set_xticklabels(products, rotation='vertical', size=8)

fig.show()
```
